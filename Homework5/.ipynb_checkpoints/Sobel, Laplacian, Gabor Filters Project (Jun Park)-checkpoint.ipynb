# Problem 1a

```python
import matplotlib.pyplot as ax 
import numpy as np 
import math 
  
n = 75
x = np.linspace(-5, 5, n)
y = 1/(1 + np.exp(-2 * x))
dx = np.gradient(y)

ax.plot(x, y)
ax2 = ax.twinx()
ax2.plot(x, dx)
ax.xlabel("x") 
```

The value of the derivative is highest at the inflection point of the sigmoid function. It is the lowest at the beginning and the end of the sigmoid function. Since the sigmoid curve is an approximation of an edge in 1-D, the derivative can tell us where the step discontinuity occurs and where the edge is essentially being generated. Thus, the point at which the derivative is the highest corresponds to the location of our edge.

# Problem 1b

```python
ndx = dx.copy()

for i in range(len(ndx)):
    if i == 0:
        ndx[i] = 0
    elif i == len(ndx) - 1:
        ndx[i] = 0
    elif dx[i] >= dx[i+1] and dx[i] >= dx[i-1]:
        ndx[i] = 1
    else:
        ndx[i] = 0

ax.plot(x, dx)
ax2 = ax.twinx()
ax2.plot(x, ndx)
ax.xlabel("x") 
```

# Problem 1c

```python
xs = np.arange(1,n-1)
dxpeaks = (dx[xs-1] <= dx[xs]) & (dx[xs] > dx[xs+1])
dxpeaks = np.pad(dxpeaks, (1, 1), 'constant', constant_values=(0, 0))

ax.plot(x, dx)
ax2 = ax.twinx()
ax2.plot(x, dxpeaks)
ax.xlabel("x") 
```

# Problem 1d

```python
ddx = np.gradient(dx)

ax.plot(x, dx)
ax2 = ax.twinx()
ax2.plot(x, ddx)
ax.xlabel("x") 
```

The point where the second derivative crosses the zero value corresponds to the point which produces the maximum of the first derivative (and also corresponds to the inflection point in the original function). Thus, the point at which the second derivative crosses zero corresponds to where our edge occurs.

# Problem 1e

```python
xs = np.arange(0, n - 1)
ddxpeaks = (ddx[xs] >= 0) & (ddx[xs+1] < 0)
ddxpeaks = np.pad(ddxpeaks, (0, 1), 'constant')

ax.plot(x, ddx)
ax2 = ax.twinx()
ax2.plot(x, ddxpeaks)
ax.xlabel("x") 
```

Since our previous procedure sets everything except the point with a value greater than or equal to both of its neighbors' values to zero, the analogous procedure for the second derivative would be to output an array containing all zeros except at the point at which the second derivative crosses 0. (Notice that our procedure ensures that we pick the point at which the derivative *crosses* 0 instead of *reaches* zero by ensuring that the point after the designated point is less than zero. Otherwise, we might end up picking the end points of the second derivative which asymptotically approach 0)

# Problem 2a

```python
from PIL import Image
import scipy
from scipy import signal

sx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sy = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

sx = sx / 8
sy = sy / 8

img_file = 'Paolina.tiff'
im = np.array(Image.open(img_file))
im = im.astype('float')/255

x = signal.convolve2d(im,sx,'same')
y = signal.convolve2d(im,sy,'same')

ax.subplot(1, 2, 1)
ax.imshow(x, cmap = "gray")
ax.subplot(1, 2, 2)
ax.imshow(y, cmap = "gray")
```

We are separately detecting edges in the x and y directions. Since the Sobel operator approximates Gaussian weighting, the application of the Sobel operator to an image is equivalent to using the sigmoid and its first derivative (which is analogous to a Gaussian) in order to detect edges in both x and y directions. More specifically, each Sobel operator is composed of three sigmoids that have been combined into a matrix. By taking their partial derivatives, we can identify the point of highest difference or change in intensity in each given direction, thereby identifying the location of the edge.

# Problem 2b

```python
gmim = np.sqrt(np.square(x) + np.square(y))
gmean = np.mean(gmim)

gmim[gmim > 2 * gmean] = 1
gmim[gmim <= 2 * gmean] = 0

ax.imshow(gmim, cmap="gray")
```

# Problem 2c

```python
def thinedge(im):
    m, n = im.shape; xs = np.arange(1,n-1); ys = np.arange(1,m-1);
    xpeaks = (im[ys][:,xs-1] <= im[ys][:,xs]) & (im[ys][:,xs] > im[ys][:,xs+1])
    ypeaks = (im[ys-1][:,xs] <= im[ys][:,xs]) & (im[ys][:,xs] > im[ys+1][:,xs])
    im = im[ys][:,xs].astype(bool) & ( xpeaks | ypeaks )
    return im

thinned = thinedge(gmim)
ax.imshow(thinned, cmap="gray")
```

As the name of the process suggests, the edges of this image have been thinned compared to those of the previous image. The process is similar to 1b-c because it selects values of our "cleaned up" image originally containing the magnitudes of the gradient that are local maxima in either the x or y direction. This process is necessary to get an edge map because it gives us a more precise estimate of the locations of our edges by removing dense lines of high intensity in our previous image that disguise where exactly edges begin and end. Edge thinning allows us to pick out the exact location of the edges (i.e. by picking out the inflection point) and is analogous to what we did in 1b-c.

# Problem 2f

We know that from 1e that the point at which the the second derivative of the sigmoid function crosses zero is the point at which an edge occurs. Therefore, by applying the Laplacian operator to an image, we are locating points where the "second derivative" is zero which corresponds to points where changes in intensity in our greyscale image is the greatest (i.e. where the edges occur). By composing a Gaussian with the Laplacian, we can think of the resulting filter as smoothing out the image before applying the Laplacian. Since the process of using second derivatives to detect edges is highly sensitive to noise, the Gaussian will stabilize the image and reduce the potential  risks of applying the Laplacian.

# Problem 3a

```python
def gaborfilter(theta, wavelength, phase, sigma, aspect, ksize=None):
    """
    GB = GABORFILTER(THETA, WAVELENGTH, PHASE, SIGMA, ASPECT, KSIZE)
    creates a Gabor filter GB with orientation THETA (in radians),
    wavelength WAVELENGTH (in pixels), phase offset PHASE (in radians),
    envelope standard deviation SIGMA, aspect ratio ASPECT, and dimensions
    KSIZE x KSIZE. KSIZE is an optional parameter, and if omitted default
    dimensions are selected.
    """

    if ksize is None:
        ksize = 8*sigma*aspect
    if type(ksize) == int or len(ksize) == 1:
        ksize = [ksize, ksize]
    
    xmax = np.floor(ksize[1]/2.)
    xmin = -xmax
    ymax = np.floor(ksize[0]/2.)
    ymin = -ymax

    xs, ys = np.meshgrid(np.arange(xmin,xmax+1), np.arange(ymax,ymin-1,-1))

    x_t = xs * np.cos(theta) + ys * np.sin(theta)
    y_t = -xs * np.sin(theta) + ys * np.cos(theta)
    
    gb = np.sin(2*np.pi*y_t/wavelength + phase) * np.exp(-(np.square(x_t)/np.square(aspect) + np.square(y_t))/(2*np.square(sigma)))
    gb= gb - np.mean(gb)
    gb = gb/np.sum(np.square(gb))
    
    return gb
```

```python
def dis_gb_filters(trange, phase):
    filts = []
    for i, theta in enumerate(np.arange(0, trange, np.pi/4)):
        gb = gaborfilter(theta, 12, phase, 3, 2)
        ax.subplot(2, 4, i+1)
        ax.imshow(gb,cmap="gray")
        filts.append(gb)
    return np.array(filts)
```

```python
# display edge filters
edgefilts = dis_gb_filters(8*np.pi/4, 0)
```

```python
# display bright line filters
blfilts = dis_gb_filters(4*np.pi/4, np.pi/2)
```

```python
# display dark line filters
dlfilts = dis_gb_filters(4*np.pi/4, 3*np.pi/2)
```

```python
# display paolina with all sixteen filters
all_filts = np.concatenate((edgefilts, blfilts, dlfilts), axis=0)
ax.figure(figsize=(15, 15))

for i in range(all_filts.shape[0]):
    filtered = scipy.signal.convolve2d(im, all_filts[i],'same')
    ax.subplot(4, 4, i+1)
    ax.imshow(filtered, vmin=0, vmax=1, cmap="gray")
```

# Problem 3c

Gabor filters are related to the receptive fields of the primary visual cortex because it allows us to have a comprehensive map of edges of all orientations. The edge map in part 2 only allows us to detect edges in the x and y dimensions. The main advantage of having Gabor filters is that we are able to directly perceive edges in all orientations instead of trying to ascertain edges of non horizontal and non vertical orientations through projections on an edge map that supports only 2 dimensions. Another advantage of Gabor filters is that they allow us to perceive edges based on other qualities of the images. For example, in Problem 3, we explored Gabor filters that are able to highlight bright or dark lines of an image. By comprehensively covering a multitude of orientations and qualities, Gabor filters allow our primary visual cortex to detect edges and objects in many different scenarios and despite unexpected changes in our environment.

```python
from mpl_toolkits.mplot3d import Axes3D

img_fft = np.fft.fftshift(np.fft.fft2(all_filts[0]))
x = np.arange(0, len(all_filts[0]))
y = np.arange(0, len(all_filts[0]))
x, y = np.meshgrid(x,y)

fig = ax.figure()
plt = fig.add_subplot(111, projection='3d')
plt.plot_surface(x,y,np.absolute(img_fft))
```

We know by the convolution theorem that the fourier transform of the pointwise product of two functions is equal to the convolution of each functions respective fourier transform. Since the Gabor filter is the pointwise product of a Gaussian and a sinusoid in the space domain, it is equivalent to the convolution of the fourier transforms of a Gaussian and a sinusoid in the frequency domain (i.e. the covolution of Gaussian (with inverse bandwidth parameter) and a dirac delta function). Thus, since the sinusoid in the space domain is a dirac delta function with two peaks, its convolution with a Gaussian would result in a bimodal function as shown above.
