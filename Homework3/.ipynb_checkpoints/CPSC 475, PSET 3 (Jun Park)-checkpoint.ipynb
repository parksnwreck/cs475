# Problem 5a

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def blurfilter(sigma, n=None):
	"""BLURFILTER Creates a Gaussian blur filter.
   FILT = BLURFILTER(SIGMA, N) creates a Gaussian blur filter with
   bandwidth parameter SIGMA, and size N x N. If N isn't specified, it
   defaults to ceil(6*sigma) + 1.

   The output of this function is the filter FILT."""

	if n == None:
	 	n = int(np.ceil(6*sigma)) + 1


	rad = (n - 1)/2.
	xs, ys = np.meshgrid(np.linspace(-rad, rad, n), np.linspace(rad, -rad, n))

	# Your code here -- (make sure you understand what xs and ys represent!)

	FILT = np.exp(-(np.square(xs)+np.square(ys))/(2 * np.square(sigma)))
	FILT = FILT/np.sum(FILT)
	return FILT
```

It is important to normalize the filter above because the goal of the filter is solely to blur the image without changing other aspects of the original image. Without normalization, there might be unexpected changes to the image that would result in adjustments to the image that are beyond the scope of the blur filter (i.e. color, brightness, and other image aspects might be altered).

# Problem 5c

```python
from PIL import Image
from scipy import signal

image_file = 'Paolina.tiff'
im = np.array(Image.open(image_file))
im = im.astype('float')/255
plt.imshow(im, cmap='gray');
plt.show()

FILT = blurfilter(3)
blurredim = signal.convolve2d(im, FILT, mode="same")
plt.imshow(blurredim, cmap='gray');
```

```python
imfft2 = np.fft.fft2(im)
blurfft2 = np.fft.fft2(FILT, im.shape)

result = np.fft.ifft2(imfft2*blurfft2).real
result = result.astype('float')

plt.imshow(result, cmap="gray")
```

The image above seems to closely resemble the 2nd image of Paolina in our sequence. The result was exspected because it is an application of our proof in 4a. Since the 2nd image of the Paolina is just a convolution between the image and the blur filter, it follows from 4a that the inverse Fourier transform of the pointwise product between the Fourier transforms of Paolina and the blur filter should result in an image that is essentially the same as the 2nd image.

The fourier transform above is considered a band pass filter because it zeroes frequencies, or in the case of an image, pixels that are not within a certain range. As shown above, there are regions that are zeroed (the valley in the middle of the plot, and the edges of the plot) while only a certain range of frequencies are accentuated (volcano-shaped mountain area). By subtracting one blur filter from another, this transformation is essentially only conserving the overlapping range between the two filters.

# Problem 5f

```python
def unsharpmask(img, sigma, amount):
    blurfilt = blurfilter(sigma)
    blurredim = signal.convolve2d(img, blurfilt, mode='same')
    
    mask = np.subtract(img, blurredim)
    unsharpened = img + amount * mask
    
    return unsharpened

image_file = 'Paolina.tiff'
im = np.array(Image.open(image_file))
im = im.astype('float')/255
plt.imshow(im, cmap='gray');
plt.show()

filteredim = unsharpmask(im, 1, 1)
plt.imshow(filteredim, cmap='gray', vmin= 0, vmax= 1);
```

The unsharp mask filter is sharpening the image by applying a mask that is a result of the original image with the blurriness removed (the effects of which are scaled by "amount" argument). This mask which is added back to the orignal image results in a sharpening effect and a reduction in overall blurriness.
