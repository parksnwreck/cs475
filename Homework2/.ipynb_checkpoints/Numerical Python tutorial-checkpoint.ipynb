## Arrays: vectors and matrices

Vectors and matrices are represented in `numpy` as arrays (type `ndarray`). As you will see, they are the building blocks of more efficient scripts, doing away with the need of `for`-loops in many cases.

```python
v = np.array([1, 4, 7, 10, 13])
type(v)
```

The first element an array:

_Note: this is a major difference from MATLAB, where the index of an array starts at 1._

```python
v[0]
```

The second up to the fourth:

```python
v[1:4] #this is called "slicing"
```

All elements from the third to the last element:    

```python
v[2:]
```

The last:

```python
v[-1]
```

```python
np.arange(10) #numpy's arange automatically creates a numpy array with the chosen range
```

```python
#not this is similar to Python's native range, but more versatile as it can handle floats
np.arange(10.)
```

```python
np.arange(1,2,.1) #range between 1 and 5 with 0.1 step
```

One thing to keep in mind, though, is that by default a vector array is not a column nor a row! You need to explicitly make it "two-dimensional" first in order to specify that. This is another major difference between `numpy` and MATLAB. Let's see what the number of dimensions and shape of our array, `v`, are:

```python
v.ndim
```

```python
v.shape #that's a funny-looking shape! not a column, not a row either.
#It just tell us the "length" of it, since it's one-dimensional
```

Using `reshape`:

```python
#column vector
v = v.reshape(5,1)
v
```

```python
v.shape #this tuple informs the number of elements in each axis
```

```python
v.ndim
```

```python
#row vector -- can apply the reshape method directly to the array definition
v = np.array([1, 4, 7, 10, 13]).reshape(1,5)
v
```

```python
v.shape
```

```python
v.ndim
```

Using the `None` slicing syntax to create an axis of length one:

```python
np.array([1, 4, 7, 10, 13])[:,None] #column
```

```python
np.array([1, 4, 7, 10, 13])[None,:] #row
```

## Array operations

You can directly add, subtract, and (elementwise) multiply two arrays that have the same shape together. Scalar addition and multiplication also applies automatically to the entire array â€“ no loops necessary!

```python
u = np.array([1, 2, 3, 4])
v = np.ones(4)

u + v
```

```python
v - u
```

```python
u * v
```

```python
v / u
```

```python
10*v
```

```python
np.square(u)
```

_Broadcasting_: In the example below, array `A` is 3-by-4. Adding `u` to it works because `numpy` automatically adds it to each row of `A`. (Also applies to multiplication.) This is a powerful funcionality but can very often be the source of bugs! 

```python
A = 2*np.ones((3,4))
A + u
```

To see how things can get easily confusing, if `B` is a 4-by-4 array, how do we know whether `u` will be added to its rows or columns? Both are in principle possible. Answer: it depends on the shape of `u`:

```python
B = np.ones((4,4))
u = np.array([1, 2, 3, 4])
u.shape
```

```python
B + u #if u is a 1-dim array, it will be added to the rows by default
```

```python
u = u.reshape(1,4) #same is true if u is a row vector
B + u
```

```python
u = u.reshape(4,1) #added to the cols is u is a column vector
B + u
```

Bottom line: it's always a good idea to keep your vectors as 2-dim arrays (explicit row- or column-vectos)!

__Matrix multiplication__

Main thing to keep in mind is whether you want matrix multiplication or elementwise (a.k.a. Hadamard) product.

```python
C = np.arange(16).reshape(4,4)
C * B #elementwise multiplication
```

```python
np.dot(C,B) #matrix multiplication
```

_Hint:_ If using Python 3.5+, you can simply do `C @ B`!

```python
np.dot(A,B) # this works because the dimensions of A (3-by-4) and B (4-by-4) are compatible
```

```python
#...but entrywise multiplication naturally won't:
A * B
```

```python
# Elementwise division also work
C / B
```

```python
# Elementwise power, too
np.power(B, C)
```

```python
# Inverse of a matrix
D = np.random.rand(3,3)
np.linalg.inv(D)
```

Dot product will work with 1-dim vector arrays but a cross product requires 2-dim vector arrays:

```python
u = np.arange(4)
np.dot(u,u)
```

```python
np.dot(u[:,None],u[None,:]) #cross product: 4-by-1 column times a 1-by-4 row gives a 4-by-4 matrix! 
```

## Logical arrays

Logical and comparison operators can also be applied to arrays in an elementwise fashion:

```python
v = np.tile([2,0],(1,3)) #get multiple copies of the array [2,0] in the shape (1,3) (i.e. 3 copies horizontally)
v
```

```python
v > 1
```

```python
v < 2
```

```python
v <= 2
```

```python
~(v > 0) #negation -- make sure to apply it to a logical array
```

```python
A = np.arange(9).reshape(3,3)
A
```

```python
# Check which elements of A are even
A % 2 == 0
```

```python
# Check which elements of A are even AND greater than 0
(A % 2 == 0) & (A > 0)
```

```python
# Check which elements of A are divisible by 3 OR divisible by 4
(A % 3 == 0) | (A % 4 == 0)
```

## Images

Read image into a 2-D matrix using the Pillow package:

```python
from PIL import Image

image_file = 'cameraman.png'
im = np.array(Image.open(image_file)) #this converts the image into a numpy ndarray
im.shape #always check the shape to see if it's a 2-D (grayscale image) or 3-D array (RGB color image)
```

Now you can use any `numpy` functions on the array. E.g., let's look at the minimum and maximum pixel values:

```python
im.min(), im.max()
```

To convert image to double precision with pixels in $[0,1]$,  change the data type of the array to floats and divide by 255 (that's because grayscale values are in the range 0-255):

```python
im = im.astype('float')/255
im.min(), im.max() #now recompute min and max vals
```

```python
#Show the grayscale image
plt.imshow(im, cmap='gray'); #set colormap as 'gray'
```

```python
# Show the heatmap of the image using the 'viridis' heatmap (usually the default one)
plt.imshow(im, cmap='viridis');
```

__Important:__ `imshow` automatically adjusts the color scale according to the min and max values.

If you want a specific range to map to the entire colormap range, you can set that with the `vmin` and `vmax` keyword arguments.

To see an example where this might be relevant, let's scale all pixel values by 0.5 and see what it does to the image:

```python
plt.imshow(0.5*im, cmap='gray')
plt.xticks([]); plt.yticks([]); #don't show axis ticks
```

You see no change because the colormap was automatically scaled to map to the new range of pixel values!

We can fix that by telling `imshow` that the full pixel value range should go from 0 (black) to 1 (white):

```python
plt.imshow(0.5*im, cmap='gray', vmin=0, vmax=1); #now you can see what scaling by a half does to the image
plt.xticks([]); plt.yticks([]); #don't show axis ticks
```

## Visualizing a matrix as an image

```python
# Generate 100 normally-distributed random points in R^2
points = np.random.randn(100,2) #put them in a 100-by-2 matrix where each row is a point
#Let's make a scatter plot to see what they look like:
plt.scatter(points[:,0],points[:,1], s=10); #s argument sets the size of the dots
```

```python
#Compute distance matrix - borrow a couple of functions from scipy first
from scipy.spatial.distance import pdist,squareform
A = pdist(points) #compute pairwise distances between all points
A = squareform(A) #convert list of distances into a square matrix (symmetric -- why?) 
# Now show matrix as an image
plt.imshow(A)
plt.colorbar();
```

```python
#Plot both figures together using subplots -- without using axes
plt.figure()
plt.subplot(121) #getting each subplot axis allows for more control over each subplot
plt.imshow(A)
plt.title('Distance matrix')
plt.subplot(122)
plt.title('Points')
plt.scatter(points[:,0],points[:,1], s=10)
plt.axis('square');
```

## Images as surfaces

```python
# Let's plot a 2-D function as a surface in a 3-D plot
%matplotlib inline
xs, ys = np.meshgrid(np.arange(1,10.5,.5),np.arange(1,21)) #create a grid of (x,y) points
zs = np.sin(xs+ys) + np.cos(ys); #compute z values at each point

#you will need to import some 3-D plotting machinery for this
from mpl_toolkits.mplot3d import Axes3D
#Then:
fig = plt.figure() #1 - get a new figure to plot on
ax = fig.add_subplot(111,projection='3d') # 2- get an axis for that figure with 3-D projection
surf = ax.plot_surface(xs, ys, zs, cmap='viridis')
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=10)
# Label each axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
#Use different colormaps in different subplots
fig = plt.figure(figsize=(4,6)) #1 - get a new figure to plot on
ax1 = fig.add_subplot(211,projection='3d') # 2- get an axis for that figure with 3-D projection
ax1.plot_surface(xs, ys, zs, cmap='gray')
ax2 = fig.add_subplot(212,projection='3d') # 2- get an axis for that figure with 3-D projection
ax2.plot_surface(xs, ys, zs, cmap='coolwarm')
```

```python
% What happens if we use surf on an image?
fig = plt.figure(figsize=(4,6)) #1 - get a new figure to plot on
ax1 = fig.add_subplot(211,projection='3d') # 2- create an axis for that figure with 3-D projection
ax1.plot_surface(xs, ys, zs, cmap='gray')
```

## Fancy indexing

You can use a vector as indices to another vector:

```python
# Let's create a fresh new array w
w = np.arange(2,13,2)
w
```

```python
# Choose some indices
idxs = range(1,6,2) #the indices don't have to be in a numpy array -- any list or tuple will do
idxs #this means we want to select entries with indices 1, 3, and 5
```

```python
w[idxs]
```

Another trick is using a logical array to select entries from another array:

```python
A = np.reshape(np.arange(9),(3,-1))
#(we can use -1 as the length of one of the axes in reshape to let numpy automatically figure it out)
A
```

```python
# Selecting the entries along the diagonal:
A[np.eye(3)>0]
```
